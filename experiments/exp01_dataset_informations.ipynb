```python
import sys
sys.path.append('../src')
import json
from mainController import MainController
from MapTypes import MapTypes
from scipy.stats import pearsonr

# Load the JSON data from a file
with open('../data/output_rbf_40_60.json') as f:
    json_data = json.load(f)

mainController = MainController(json_data)
total_features = len(json_data['features'])

print(mainController.dto.gdf.columns)
print(f"total_features: {total_features}")
```

```python
total_all_measurements = mainController.dto.gdf['all_measurements'].sum()
print(f"total_all_measurements: {total_all_measurements}")
```

```python
total_measurements_tmobile = mainController.dto.gdf['t-mobile_measurements'].sum()
total_measurements_vodafone = mainController.dto.gdf['vodafone_measurements'].sum()
total_measurements_o2 = mainController.dto.gdf['e-plus_measurements'].sum()
total_measurements_eplus = mainController.dto.gdf['o2_measurements'].sum()

print(f"total_measurements t-mobile: {total_measurements_tmobile:<10}")
print(f"total_measurements vodafone: {total_measurements_vodafone:<10}")
print(f"total_measurements o2: {total_measurements_o2:<10}")
print(f"total_all_measurements e-plus: {total_measurements_eplus:<10}")
```

```python
import matplotlib.pyplot as plt

providers = ['T-Mobile', 'Vodafone', 'O2', 'E-Plus']
measurements = [total_measurements_tmobile, total_measurements_vodafone, total_measurements_o2, total_measurements_eplus]

# Colorblind-friendly colors
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# Create the plot
plt.bar(providers, measurements, color=colors)
plt.xlabel('Provider')
plt.ylabel('Number of Measurements')
plt.title('Number of Measurements by Provider')
# Format y-axis labels in a readable way
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```
